```python
# Part 1: Importing libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# Function to print scanning message
def print_scanning_message(page_num):
    print(f"Scanning page {page_num}... Time: {time.strftime('%H:%M:%S')}")
```

```python
# Function to traverse through a range of pages
def traverse_pages(start_page, end_page):
    all_data = []
    for page_num in range(start_page, end_page + 1):
        print_scanning_message(page_num)
        url = f"https://seriesonline.ac/movie?page={page_num}"
        movie_data = scrape_movie_data(url)
        all_data.extend(movie_data)
    return all_data
```

```python
# Part 2: Scraping function
def scrape_movie_data(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    data = []
    all_movies = soup.find_all("div", class_="film-detail")
    for movie in all_movies:
        item = {}
        item['Title'] = movie.find("h2", class_="film-name").text.strip()
        item['Year'] = movie.find("span", class_="fdi-item").text.strip()
        item['Duration'] = movie.find("span", class_="fdi-item fdi-duration").text.strip()
        data.append(item)
    return data
```

```python
# Part 3: Execution
start_page = int(input('Enter Start page:'))
end_page = int(input('Enter End page:'))
movie_data = traverse_pages(start_page, end_page)
df = pd.DataFrame(movie_data, columns=['Title', 'Year', 'Duration'])
print(df)

df.to_csv("movies.csv")
```
